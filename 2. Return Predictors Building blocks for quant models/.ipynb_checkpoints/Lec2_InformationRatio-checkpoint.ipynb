Computes Sharpe ratio for long and Long/Short portfolios.




**Data Description**

Important Dataframes

1.  "Returns" dataframe : Input time series of portfolio returns.

Input file : "b2m_monthlyReturns.csv", created by "Lec2_B2m.ipynb"


2.   "Risk_freeRisk_freeRate_market" dataframe : Input time series of risk-free rate + market Returns

Market = Market-cap weighted index of all US stocks from CRSP (vwretd)

Input file : "Market_Riskfree.csv"








```python
# Importing Necessary Python Libraries
import pandas as pd
import numpy as np
```

```python
# Import Portfolio Returns Data
Returns = pd.read_csv("/content/drive/MyDrive/MAF data/b2m_monthlyReturns.csv") #b2m decile portfolio returns created by Lec2_B2m.ipynb

# Risk-Free rate
# Importing market return and risk free rate data. "Market" is the return on market cap-weighted return of all stocks on US exchanges, and "RiskfreeRate" is the return on one-month Treasury Bill
Risk_freeRate_market = pd.read_csv('/content/drive/MyDrive/MAF data/Market_Riskfree_2312.csv')     # read risk-free rate and market returns
merge_data = pd.merge(Returns, Risk_freeRate_market, how = "inner", on = ['year', 'month'])
merge_data
```

```python
#Information ratio = (portfolio return - market returns)/tracking error
merge_data['rp minus rm'] = merge_data['RET'] - merge_data['Market']
ranks = list(np.unique(merge_data['rank']))
ranks
```

```python
# Decile portfolio return means and annualized standard deviation
for rank in ranks:
  df = merge_data.loc[merge_data['rank'] == rank]
  print('rank=', rank, 'mean =', "{:.4f}".format(df['RET'].mean()), 'standard deviation_annual=', "{:.4f}".format(np.sqrt(12) * df['RET'].std()))
```

Q. Intuitively, why does the standard deviation have an u-shaped pattern across deciles?

```python
np.sqrt(12) * merge_data['Market'].std()
```

```python
# Decile portfolio information ratio and annualized tracking error
for rank in ranks:
  df = merge_data.loc[merge_data['rank'] == rank]
  tracking_error = df['rp minus rm'].std()
  Information_ratio_annual = np.sqrt(12) * df['rp minus rm'].mean()/tracking_error
  print('rank=', rank, 'Information_ratio_annual= ', "{:.4f}".format(Information_ratio_annual), 'tracking error_annual=', "{:.4f}".format(np.sqrt(12) * tracking_error) )
```
